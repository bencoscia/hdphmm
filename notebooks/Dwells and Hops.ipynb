```python
nboot = 200

confidence = 95

dwells_ = []
hops_ = []

dwells_range = (1, 150)
hops_range = (-2, 2)

bins_dwell = 25
bins_hop = 51

traj_no = np.arange(len(ihmm))

for t in traj_no:
    
    zeroed, d, h = ihmm[t].subtract_mean(traj_no=0, simple_mean=True, return_dwells_hops=True)
    dwells_.append(d)
    hops_.append(np.array(h)[:, 2].tolist())

hops, hop_limits = boot(hops_, nboot, bins_hop, hops_range, confidence)
dwells, dwell_limits = boot(dwells_, nboot, bins_dwell, dwells_range, confidence)
pre_cluster_hops, pre_cluster_hops_limits = boot(pre_cluster_hops_, nboot, bins_hop, hops_range, confidence)
pre_cluster_dwells, pre_cluster_dwells_limits = boot(pre_cluster_dwells_, nboot, bins_dwell, dwells_range, confidence)

edges = np.linspace(dwells_range[0], dwells_range[1], bins_dwell + 1)
edges_hop = np.linspace(hops_range[0], hops_range[1], bins_hop + 1)
bin_width = edges[1] - edges[0]
bin_width_hop = edges_hop[1] - edges_hop[0]
centers_dwell = [i + bin_width/2 for i in edges[:-1]]
centers_hop = [i + bin_width_hop/2 for i in edges_hop[:-1]]
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].plot(centers_dwell, dwells, lw=2, label='MD Data')
ax[0].fill_between(centers_dwell, dwells + dwell_limits[0, :], dwells - dwell_limits[1, :], alpha=0.3)#, color='xkcd:black')

ax[0].plot(centers_dwell, pre_cluster_dwells, lw=2, label='IHMM Simulation Data')
ax[0].fill_between(centers_dwell, pre_cluster_dwells + pre_cluster_dwells_limits[0, :], pre_cluster_dwells - pre_cluster_dwells_limits[1, :], alpha=0.3)#, color='xkcd:black')

ax[1].plot(centers_hop, hops, lw=2, label='MD Data')
ax[1].fill_between(centers_hop, hops + hop_limits[0, :], hops - hop_limits[1, :], alpha=0.3)

ax[1].plot(centers_hop, pre_cluster_hops, lw=2, label='IHMM Simulation Data')
ax[1].fill_between(centers_hop, pre_cluster_hops + pre_cluster_hops_limits[0, :], pre_cluster_hops - pre_cluster_hops_limits[1, :], alpha=0.3)

ax[0].set_xlabel('Dwell Time (steps)', fontsize=14)
ax[0].set_ylabel('Probability', fontsize=14)
ax[0].tick_params(labelsize=14)
ax[0].legend(fontsize=14)

# file_rw.save_object({'hop_centers': centers_hop, 'dwell_centers': centers, 'x': x, 'levy': levy, 'normal': normal, 
#                      'raw_hops': pre_cluster_hops, 'raw_dwells': pre_cluster_dwells, 'MD_dwells': dwells, 
#                      'MD_hops': hops}, 'dwell_hops_ar%d.pl' % ihmm[0].order)

ax[1].set_xlabel('Jump length (nm)', fontsize=14)
ax[1].set_ylabel('Probability', fontsize=14)
ax[1].tick_params(labelsize=14)
ax[1].legend(fontsize=14)

fig.tight_layout()
plt.show()
```
